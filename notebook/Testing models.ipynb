```python
# Notebook used to explore the data and the possible models
import pandas as pd

df = pd.read_csv('C:/Users/Alex/PycharmProjects/barcelona-apartments-2/data/processed/barcelona_apartments2.csv')
```

```python
# Create a onehot encoder category for the districts in order to do stratified train test split
df['district_cat'] = df['eixample'].astype(str) + df['ciutat_vella'].astype(str) + df['horta_guinardo'].astype(str) + df['gracia'].astype(str) + df['les_corts'].astype(str) + df['nou_barris'].astype(str) + df['sant_andreu'].astype(str) + df['sant_marti'].astype(str) + df['sants_montjuic'].astype(str) + df['sarria_sant_gervasi'].astype(str)
```

```python
df['district_cat'].value_counts()
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.2, random_state=42, stratify=df['district_cat'])
```

```python
test['district_cat'].value_counts()
```

```python
for set_ in (train, test):
    set_.drop('district_cat', axis=1, inplace=True)
```

```python
train.corr()
```

```python
train.head(-1)
```

```python
X_train = train.drop("price", axis=1)
y_train = train["price"].copy()
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train[["rooms", "bathrooms", "sizem2"]] = scaler.fit_transform(X_train[["rooms", "bathrooms", "sizem2"]])
```

```python
X_train.head()
```

```python
import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()

scores = cross_val_score(lin_reg, X_train, y_train, scoring="neg_mean_squared_error", cv=5)
tree_rmse_scores = np.sqrt(-scores)
tree_rmse_scores.mean()
```

```python
lin_reg.fit(X_train, y_train)
predictions = lin_reg.predict(X_train)
mean_squared_error(predictions, y_train, squared=False)
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2, include_bias=False)

X_poly = poly.fit_transform(X_train)
```

```python
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=100, solver="cholesky")

scores = cross_val_score(ridge, X_poly, y_train, scoring="neg_mean_squared_error", cv=5)
tree_rmse_scores = np.sqrt(-scores)
tree_rmse_scores.mean()
```

```python
ridge.fit(X_poly, y_train)
predictions = ridge.predict(X_poly)
mean_squared_error(predictions, y_train, squared=False)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid =[
    {'solver':["cholesky"], 'alpha':[13, 14, 15, 16, 17, 18, 19]}
]

grid_search = GridSearchCV(ridge, param_grid, cv=5, scoring='neg_mean_squared_error', return_train_score=True)

grid_search.fit(X_poly, y_train)
```

```python
results = grid_search.cv_results_
for mean, params in zip(results["mean_test_score"], results["params"]):
    print(np.sqrt(-mean), params)
```
